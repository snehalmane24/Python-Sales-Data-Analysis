# Exploratory Data Analysis[EDA]

### 1. Gender

```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df)
# show labels on bars
for bars in ax.containers:
    ax.bar_label(bars) 
```

```python
# How much amount spend by each gender and sort values by amount in ascending order.
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# How much amount spend by each gender and sort values by amount in ascending order. Represent it by Visual Format.
sales_gen =df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

#### From above graph, We can see that most of the buyers are Females and even the purchasing power of Females are greater than Males.

### 2. Age

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender') #hue : Divide a data of Gender on the basis Male & Female
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# How much amount spent by each age group and sort values by amount in ascending order.
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
# How much amount spent by each age group and sort values by amount in ascending order. Represent it in visual format.
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

#### From above group we can see that most of the buyers are of age between 26-35 years Females

### 3. State

```python
df.columns
```

```python
#Total No. Of Orders from Top 10 States
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
#Show Visually, total number of Orders from Top 10 States 
plt.figure(figsize=(16,6)) #Width,Height
sales_age=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(x='State',y='Orders',data=sales_age)
for bars in ax.containers: #Show labels on bars
    ax.bar_label(bars)
```

```python
#Total Amount from top 10 States
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
#Represent Visually, Total Amount from top 10 States.
plt.figure(figsize=(16,6)) #Width,Height
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(x='State',y='Amount',data=sales_state)
```

#### From above graph, We can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashtra and Karnataka respectively.

### 4. Marital Status

```python
df.columns
```

```python
#On the basis of count
ax=sns.countplot(x='Marital_Status',data=df)
plt.figure(figsize=(7,5))
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#On the basis of Amount
sales_status= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_status
```

```python
sales_status= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(data=sales_status,x='Marital_Status',y='Amount',hue='Gender')
```

### 5. Occupation

```python
df.columns
```

```python
plt.figure(figsize = (20,5))
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(20,5))
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Occupation',y='Amount',data=sales_occ)
for bars in ax.containers:
    ax.bar_label(bars)
```

#### From above graph, Most of the buyers working in IT Sector, Healthcare, Aviation.

### 6. Product Category

```python
df.columns
```

```python
# Countplot of product_category
plt.figure(figsize=(25,5))
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 product categories spend by highest amount
plt.figure(figsize=(20,5))
sales_procat=df.groupby('Product_Category',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(x='Product_Category',y='Amount',data=sales_procat)
```

#### From above graph, We can see that most of the sold products are from Food, Clothing & Apparel, Electronics & Gadgets.

### ProductID

```python
#Top selling product on basis of Product ID
plt.figure(figsize=(20,5))
sales_proid=df.groupby('Product_ID',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
ax=sns.barplot(x='Product_ID',y='Orders',data=sales_proid)
for bars in ax.containers:
    ax.bar_label(bars)
```

### CONCLUSION :

#### Married women age group 26-35 yrs from UP, Maharashtra & Karnataka Working in IT, Healthcare & Aviation are more likely to buy products from Food ,Clothing & Electronics Category.
